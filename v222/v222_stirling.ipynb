```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Plotten des tehrmischen und effektiven Wirkungsgrades √ºber die Frequenz

#Werte zusammenfassen in Arrays: 
x = np.array([3.02, 3.88, 4.21, 4.82]) #Motor-Frequenzen
dx = np.array([0.05, 0.15, 0.12, 0.04]) #Fehler Motorfrequenz

y1 = np.array([0.025, 0.024, 0.018, 0.010]) #effektiver Wirkungsgrad
dy1 = np.array([0.002, 0.002, 0.002, 0.003]) #Fehler eff Wirk

y2 = np.array([0.058, 0.071, 0.074, 0.082]) #thermischer Wirkungsgrad
dy2 = np.array([0.001, 0.003, 0.002, 0.001]) #Fehler therm Wirk

plt.errorbar(x, y2, xerr=dx, yerr=dy2, linestyle='-', marker='x', label='$Thermischer\ Wirkungsgrad\ \eta_{th}$', color='c')
plt.errorbar(x, y1, xerr=dx, yerr=dy1, linestyle='-', marker='x', label='$Effektiver\ Wirkungsgrad\ \eta_{eff}$')
plt.xlabel('$Hei√üluftmotorfrequenz\ f\ in\ [\\frac{1}{s}]$')
plt.ylabel('$Wirkungsgrad $')
plt.title(r'Diagramm 11: Effektiver und Thermischer Wirkungsgrad')
#plt.yscale('log')
plt.grid(True)
plt.axis((2.8, 5, 0, 0.09))
plt.rcParams["figure.figsize"][0] = 8
plt.rcParams["figure.figsize"][1] = 5
plt.rcParams["font.family"]='serif'
plt.legend(title='Legende: Messwerte', borderpad=0.7, borderaxespad=1, loc='best', shadow='true')

#plt.savefig('v222_1.pdf',format='pdf')
```
