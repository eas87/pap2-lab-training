Aufgabe 1a: Gleichspannungsverstärkung

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


U_ein, U_aus1, U_aus2 = np.loadtxt('tab1.txt', skiprows=1, usecols=(0,1,2), unpack=True)

plt.plot(U_ein, U_aus1, linestyle='None', marker='.', label='$R_{G}=48.7 k\Omega$')
plt.plot(U_ein, U_aus2, linestyle='None', marker='x', label='$R_{G}=274 k\Omega$')
plt.xlabel('Eingangsspannung $U_{ein}$')
plt.ylabel('Ausgangsspannung $U_{aus}$')
plt.title('Spannungsverstaerkung: $U_{aus}=f(U_{ein})$ bei DC (Tabelle1)')
plt.legend(title='Gegenkoppelwiderstand', loc='best', shadow='true')
```

Aufgabe 1b: Wechselspannungsverstärkung

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


U_ein, U_aus1, U_aus2, U_aus3 = np.loadtxt('tab2.txt', skiprows=1, usecols=(0,1,2,3), unpack=True)

U_ein = U_ein/10

plt.plot(U_ein, U_aus1, linestyle='None', marker='.', label='$R_{G}=48.7 k\Omega$')
plt.plot(U_ein, U_aus2, linestyle='None', marker='x', label='$R_{G}=274 k\Omega$')
plt.plot(U_ein, U_aus3, linestyle='None', marker='o', label='$R_{G}=680 k\Omega$')
plt.xlabel('Eingangsspannung $U_{ein}$')
plt.ylabel('Ausgangsspannung $U_{aus}$')
plt.title('Spannungsverstaerkung: $U_{aus}=f(U_{G}/10)$ bei AC (Tabelle2)')
plt.legend(title='Gegenkoppelwiderstand', loc='best', shadow='true')
```

Aufgabe 2: Frequenzgang mit Gegenkopplungen und Kondensator

```python
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print "Current size:", fig_size
 
# Set figure width to 12 and height to 9
fig_size[0] = 10
fig_size[1] = 7
plt.rcParams["figure.figsize"] = fig_size

f, U_A1, U_A2, U_A3, U_A4, U_A5 = np.loadtxt('tab3.txt', skiprows=1, usecols=(0,1,2,3,4,5), unpack=True)


plt.loglog(f, U_A1, linestyle='None', marker='.', label='$R_{G}=680 k\Omega, U_{G}=0.3 V_{SS}$')
plt.loglog(f, U_A2, linestyle='None', marker='x', label='$R_{G}=274 k\Omega, U_{G}=0.3 V_{SS}$ ')
plt.loglog(f, U_A3, linestyle='None', marker='o', label='$R_{G}=48.7 k\Omega, U_{G}=1.0 V_{SS}$')
plt.loglog(f, U_A4, linestyle='None', marker='+', label='$R_{G}=48.7 k\Omega, U_{G}=1.0 V_{SS}, C=540 pF$')
plt.loglog(f, U_A5, linestyle='None', marker='s', label='$R_{G}=48.7 k\Omega, U_{G}=1.0 V_{SS}, C=47 nF$')
plt.xlabel('Frequenz $f$')
plt.ylabel('Ausgangsspannung $U_{aus}$')
plt.title('Aufgabe 2: Frequenzgang $V\'=f(v)$ (Tabelle3,4,5)')
plt.legend(title='Widerstand und Eingangsspannung:', loc='lower left', shadow='true')
plt.savefig('bild.png', figsize=(40,20), dpi=144)
```
